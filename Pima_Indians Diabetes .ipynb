```python
preg_mean = df["Pregnancies"].mean()
preg_samp = np.random.choice(df["Pregnancies"], 100)
```

```python
from scipy.stats import ttest_1samp
ttest, p_value = ttest_1samp(preg_samp, preg_mean)
```

```python
print("p_value for T-test on pregnancy: ", p_value)
```

```python
if p_value <= 0.05:
    print("we are REJECTING null hypothesis..!")
else:
    print("we are ACCEPTING null hypothesis..!")
```

```python
from scipy.stats import ttest_ind
def t_test(x, y):
    samp_x = np.random.choice(x, 100)
    x_mean = x.mean()
    samp_y = np.random.choice(y, 100)
    y_mean = y.mean()

    ttest, p_value = ttest_ind(samp_x, samp_y)
    if p_value <= 0.05:
        print(p_value)
        return "we are REJECTING null hypothesis..!"
    else:
        print("p_value:", p_value)
        return "we are ACCEPTING null hypothesis..!"
    
```

```python
t_test(df['BMI'], df['Glucose'])
```

```python
# defining X and Y values
x = df.drop('Outcome', axis = 1)
y = df['Outcome']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.3, random_state = 4);

reg = LogisticRegression()
reg.fit(x_train, y_train)
```

```python
coefficient = reg.coef_
intercept = reg.intercept_
print("Coefficient = ",coefficient, "intercept = ", intercept)
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
print(score)
```
